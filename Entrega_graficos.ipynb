### Dataset de clientes de banca - Matplotlib

Variables Potenciales para clientes de banca: poder saber cual es la necesidad de los productos financieros para un mercado especifico e inclusive segmentar el grupo objetivo para llegar con marketing al mismo

```python
DATA_PATH = 'BankChurners.csv'
```

```python
data = pd.read_csv(DATA_PATH)
```

```python
data
```

```python
data.columns
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
plt.yticks(range(0, 15000, 200))
data['Card_Category'].value_counts().plot(kind="bar")
ax.set_ylabel("cantidad de clientes ")
ax.set_xlabel("Tipo de tarjeta de crédito")
```

Este gráfico me permito saber cuantos clientes tengo por tipologia de tarjeta de crédito, de esta forma puedo infererir que la tarjeta que mas se adquiere es la BLUE, pero que la platinum tiene mejor retribución siendo la que menos se usa.

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
plt.scatter(data['Credit_Limit'],data['Total_Trans_Ct'])
plt.xticks(range(0, 36000, 1000))
plt.yticks(range(0, 150, 10))
ax.set_xlabel("Limites de tarjeta de crédito  (en miles de dolares)")
ax.set_ylabel("Total de Transacciones")
```

En este gráfico podemos ver que la mayor parte de transacciones estan entre 30 y 90 transacciones y que en su mayor parte tienden a ser por montos entre los 1000 y 10.000 dolares

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))

x= data.Credit_Limit
y= data.Customer_Age

plt.style.use("default")
plt.hist2d(x, y, bins = (25, 20));
plt.colorbar();
ax.set_xlabel("Limites de tarjeta de crédito  (en miles de dolares)")
ax.set_ylabel("Edad del cliente")

#ax.hist(data['Customer_Age'], bins = 10, alpha = 0.45, color = 'red')
#plt.xticks(range(25, 70, 5))
```

En estgráfico hsitograma de dos variables en a que una es la edad y la otra el limite de crédito podemos comparar que limite tienen las tarjetas de criedito por el segmento de edad, lo que nos lleva a inferir que:

1. Hay mayor cantidad de usuarios de edades entre 45 a 50 años con un limite de tarjetas de crédito inferior a los 5000 dolares.
2. Su contraparte se posiciona con la misma edad con no mayor a 50 puntos con limites en tarjetas de credito superiores a los 30.000 dolares

### Dataset de clientes de banca - SEABORN

```python
tableCross1 = pd.crosstab(data['Education_Level'], data['Marital_Status'])
```

```python
tableCross1
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))

sns.lineplot(data=tableCross1)
ax.tick_params(axis='x', rotation=60)
```

En el grafico superior vemos que predomina el cliente que es graduado y a la vez tiene un matrimonio por lo que puede este ultimo ser un influenciador para el incremento de la compra y utilización de las tarjetas de crédito, al igual que las personas que son casadas y a la vez graduadas tienen un tipo de tarjeta blue

```python
sns.boxplot(x=tableCross1["Married"])
```

Vemos la diferencia de personas que estan casadas con respecto a la cantidad de clientes que aplican en otros aspectos a las cuentas de tarjetas de crédito.

```python
tableCross2 = pd.crosstab(data['Customer_Age'], data['Card_Category'])
```

```python
tableCross2.head()
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))

sns.scatterplot(data=tableCross2, x='Blue', y='Customer_Age')
```

El uso de la tarjeta Gold tiene mucho más clientes de edades entre 40 y 50 años, aunque en comparacion tambien exite un gran porcentaje de edades entre los 65 a 74 años que no tienen tarjeta, que peuden ser un segmento de mercado al cual se le puede llegar, esto revisando si la información e historial crediticio les permite tener una tarjeta de crédito.
